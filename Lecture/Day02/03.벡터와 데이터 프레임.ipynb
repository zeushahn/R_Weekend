```r
# 벡터 계산
x <- c(3,6,9,12,15)
x + c(1,2,3,4,5)
```

```r
# 벡터 합치기 (1)

x <- c(3,5,6,12,15)
y <- c(1,2,3,4,5)
z <- append(x, y); z

length(z) == length(x) + length(y)
```

```r
# iris의 Sepal.Length와 iris의 Sepal.Width의 데이터를 Sepal로 벡터 합치고 확인
Sepal <- append(iris$Sepal.Length, iris$Sepal.Width)
length(Sepal)
```

```r
# 벡터 합치기 (2)

x <- c(3,5,6,12,15)
y <- c(1,2,3,4,5)
z <- c(x, y); z

length(z) == length(x) + length(y)
```

```r
# iris의 Sepal.Length와 iris의 Sepal.Width의 데이터를 Sepal로 벡터 합치고 확인
Sepal <- c(iris$Sepal.Length, iris$Sepal.Width)
length(Sepal)
```

```r
# iris의 Sepal.Length의 중복되지 않은 갯수의 percent 구하기

length(unique(iris$Sepal.Length)) / length(iris$Sepal.Length) * 100
```

```r
# iris의 Sepal.Length의 중복된 갯수의 percent 구하기

(1 - length(unique(iris$Sepal.Length)) / length(iris$Sepal.Length)) * 100
```
